## A . Mean, Median, Mode â€” Measures of central tendency

```python

## 1. Measures of Central Tendency
import statistics
data = [12, 15, 21, 15, 18, 22, 15, 19, 20]
# Mean
mean_value = statistics.mean(data)
## Mean is a measure of central tendency, not of spread, shape, or distribution.
## The mean tells you the mathematical center of your data, but itâ€™s most meaningful when the data is symmetrical and free of outliers.

# Median
median_value = statistics.median(data)
# The median tells you the middle value of a dataset when it's arranged in order â€” and itâ€™s especially useful for understanding the central location of data when there are outliers or skewed distributions.
# Median tells you the center of your data â€” and itâ€™s your go-to measure when the mean is misleading due to outliers or skewed data.

# Scenario	Why Median Is Useful : - 
# Income distribution	Avoids distortion from a few ultra-rich individuals
# Real estate prices	Reflects typical home value better than average
# Exam scores with outliers	Shows central tendency without being skewed by extremes

# Mode
mode_value = statistics.mode(data)
# Useful for categorical data where mean and median may not apply.
# Helps spot popular choices, trends, or dominant categories.
# Itâ€™s especially powerful for non-numeric or skewed data.
# Unlike mean and median, mode can be used for nominal data (e.g., colors, brands, categories).

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

## B. Range, Variance, Standard Deviation

```python
import statistics
data = [12, 15, 21, 15, 18, 22, 15, 19, 20]

# Range
range_value = max(data) - min(data)
## Formula: Range = Maximum value âˆ’ Minimum value
## Range is the simplest measure of spread â€” it tells you how far apart your smallest and largest values are.
## Itâ€™s useful for quick checks on variability but can be distorted by outliers.

# Scenario  Why Range Is Useful : -
# Loan amounts  Shows the gap between smallest and largest loans
# Transaction values    Highlights extreme values in customer behavior
# Credit scores Indicates diversity in borrower profiles

# Variance
variance_value = statistics.variance(data)
## Formula: Variance (sÂ²) = Î£(xáµ¢ âˆ’ xÌ„)Â² / (n âˆ’ 1)
## Where xáµ¢ = each data point, xÌ„ = mean, n = number of data points
## Variance measures how far each data point is from the mean â€” on average.
## Itâ€™s expressed in squared units, which makes it less intuitive but useful for statistical modeling.

# Scenario  Why Variance Is Useful : -
# Investment returns    Quantifies risk and volatility
# Claim amounts Detects inconsistency in insurance payouts
# Customer spending Reveals behavioral spread across segments

print("Range:", range_value)
print("Variance:", variance_value)
```

```python
# ===============================
# VARIANCE â€” Concept & Use Cases
# ===============================

import statistics

# -------------------------------
# 1ï¸âƒ£ Definition & Formula
# -------------------------------
# Variance measures how far each data point is from the mean â€” on average.
# Formula:  sÂ² = Î£(xáµ¢ âˆ’ xÌ„)Â² / (n âˆ’ 1)
# Where:
#   xáµ¢ = each data point
#   xÌ„  = mean of the dataset
#   n   = number of data points
# Itâ€™s expressed in squared units and is key in risk, reliability, and behavioral analysis.

# -------------------------------
# 2ï¸âƒ£ Scenario 1: Investment Returns â†’ Quantifies Risk and Volatility
# -------------------------------
stock_A = [5, 7, 6, 8, 6]
stock_B = [2, 10, -1, 12, 0]

var_A = statistics.variance(stock_A)
var_B = statistics.variance(stock_B)

print("ðŸ“Š Investment Returns")
print(f"Stock A Variance: {var_A:.2f}")
print(f"Stock B Variance: {var_B:.2f}")
print("Higher variance (Stock B) â‡’ More volatility and risk.\n")

# -------------------------------
# 3ï¸âƒ£ Scenario 2: Insurance Claim Amounts â†’ Detects Inconsistency
# -------------------------------
branch_X = [10000, 10500, 9800, 10200]
branch_Y = [2000, 15000, 8000, 20000]

var_X = statistics.variance(branch_X)
var_Y = statistics.variance(branch_Y)

print("ðŸ§¾ Insurance Claim Amounts")
print(f"Branch X Variance: {var_X:.2f}")
print(f"Branch Y Variance: {var_Y:.2f}")
print("Branch Y has higher variance â‡’ Greater inconsistency or irregular claims.\n")

# -------------------------------
# 4ï¸âƒ£ Scenario 3: Customer Spending â†’ Reveals Behavioral Spread
# -------------------------------
segment_A = [300, 320, 280, 310]
segment_B = [100, 500, 250, 700]

var_A_seg = statistics.variance(segment_A)
var_B_seg = statistics.variance(segment_B)

print("ðŸ›ï¸ Customer Spending Behavior")
print(f"Segment A Variance: {var_A_seg:.2f}")
print(f"Segment B Variance: {var_B_seg:.2f}")
print("Segment B has higher variance â‡’ Diverse spending habits across customers.\n")

# -------------------------------
# 5ï¸âƒ£ Summary
# -------------------------------
print("ðŸ” Summary:")
print("""
| Scenario           | What Variance Measures           | Why It Matters                     |
|--------------------|----------------------------------|------------------------------------|
| Investment Returns | Risk & volatility in returns     | Helps assess stability of portfolio|
| Insurance Claims   | Inconsistency in payouts         | Detects anomalies or irregularities|
| Customer Spending  | Behavioral diversity             | Guides segmentation & personalization|
""")
```

## C. Skewness, Kurtosis

```python
# ðŸ“Š Skewness and Kurtosis in BFSI Analytics

# Understanding the shape of data distributions is critical in financial modeling. Skewness and kurtosis help assess asymmetry and tail risk, which are vital in credit scoring, fraud detection, and portfolio analysis.



## ðŸ“ Skewness

# - **Types**:
#   - **Positive Skew**: Tail on the right (e.g., income distribution with few ultra-rich)
#   - **Negative Skew**: Tail on the left (e.g., early loan repayments)
#   - **Zero Skew**: Symmetrical distribution



## ðŸ“ Kurtosis

# - **Definition**: Kurtosis measures the "tailedness" of a distribution.
# - **Formula**:
  

# \[
#   \text{Kurtosis} = \frac{n(n+1)}{(n-1)(n-2)(n-3)} \sum \left( \frac{x_i - \bar{x}}{s} \right)^4 - \frac{3(n-1)^2}{(n-2)(n-3)}
#   \]


# - **Types**:
#   - **Leptokurtic (>3)**: Heavy tails, sharp peak (e.g., rare but extreme claims)
#   - **Platykurtic (<3)**: Light tails, flat peak (e.g., uniform transaction values)
#   - **Mesokurtic (=3)**: Normal distribution

# ---

## ðŸ’¼ BFSI Use Cases

# | Concept     | BFSI Application Example                          |
# |-------------|---------------------------------------------------|
# | Skewness    | Detecting asymmetric repayment patterns, fraud    |
# | Kurtosis    | Assessing tail risk in portfolios, claim severity |

# ---

## ðŸ“Š Python Visualization


from scipy.stats import skew, kurtosis
import seaborn as sns
import matplotlib.pyplot as plt

# Sample datasets
data_pos_skew = [2, 3, 4, 5, 6, 7, 8, 20]
data_neg_skew = [20, 18, 17, 16, 15, 14, 13, 2]
data_normal = [10, 12, 13, 14, 15, 16, 17, 18]

datasets = {
    "Positively Skewed": data_pos_skew,
    "Negatively Skewed": data_neg_skew,
    "Normal Distribution": data_normal
}

plt.figure(figsize=(15, 4))

for i, (label, values) in enumerate(datasets.items(), 1):
    plt.subplot(1, 3, i)
    sns.histplot(values, kde=True, bins=8, color='skyblue')
    sk = skew(values)
    kurt = kurtosis(values)
    plt.title(f"{label}\nSkewness: {sk:.2f}, Kurtosis: {kurt:.2f}")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### Kurtosis

```python
import numpy as np
from scipy.stats import kurtosis
import matplotlib.pyplot as plt

# Create sample data
normal_data = np.random.normal(0, 1, 1000)         # Mesokurtic
heavy_tail_data = np.random.laplace(0, 1, 1000)    # Leptokurtic
flat_data = np.random.uniform(-2, 2, 1000)         # Platykurtic

# Calculate kurtosis (Fisher=True means 'excess kurtosis')
print("Kurtosis Values:")
print(f"Normal Distribution: {kurtosis(normal_data):.2f}")
print(f"Laplace (Heavy tails): {kurtosis(heavy_tail_data):.2f}")
print(f"Uniform (Flat): {kurtosis(flat_data):.2f}")

# Plot comparison
plt.figure(figsize=(8,5))
plt.hist(normal_data, bins=30, alpha=0.5, label='Normal')
plt.hist(heavy_tail_data, bins=30, alpha=0.5, label='Leptokurtic (Laplace)')
plt.hist(flat_data, bins=30, alpha=0.5, label='Platykurtic (Uniform)')
plt.legend()
plt.title("Kurtosis Comparison")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### How to handle Skewness and Kurtosis

```python
Handling Skewness
ðŸ”„ 1. Apply Data Transformations
Used to reduce asymmetry and make data more normal-like.

Transformation	Use Case	Python Example
Log Transform	Right-skewed income or claim amounts	np.log(x)
Square Root	Moderate skew	np.sqrt(x)
Box-Cox	Flexible, auto-optimized	scipy.stats.boxcox(x)
Yeo-Johnson	Works with zero/negative values	PowerTransformer(method='yeo-johnson')
âœ… BFSI Tip: Use log or Box-Cox for skewed transaction amounts, loan sizes, or asset returns.

ðŸ§¹ 2. Remove or Cap Outliers
Outliers often cause skewness.

Winsorization: Replace extreme values with percentiles (e.g., 5th and 95th)

Trimming: Remove top/bottom x% of values

Capping: Set upper/lower bounds

âœ… BFSI Tip: Cap unusually large claims or transactions to stabilize model inputs.

ðŸ“Š 3. Use Robust Models
Some models handle skewed data better:

Tree-based models (Random Forest, XGBoost)

Quantile regression

Non-parametric methods

âœ… BFSI Tip: Use tree models for fraud detection where skewed behavior is common.

ðŸ“ˆ Handling Kurtosis
    
ðŸ§ª 1. Use Tail-Robust Metrics
High kurtosis means heavy tailsâ€”more extreme values.

Use median, IQR, or robust z-scores instead of mean/std

Apply quantile-based binning for segmentation

âœ… BFSI Tip: In credit risk, use IQR to assess borrower variability without being misled by extremes. 


ðŸ” 2. Stress Testing
If kurtosis is high, simulate tail events:

Monte Carlo simulations
Value at Risk (VaR) modeling
Scenario analysisL

âœ… BFSI Tip: Insurance and portfolio teams use this to prepare for rare but severe losses.


ðŸ§  3. Feature Engineering
Create new features that reduce tail sensitivity:

Flag extreme values

Use log ratios or percentiles

Normalize within segments

âœ… BFSI Tip: Create â€œhigh-riskâ€ flags for claim amounts or repayment delays beyond 95th percentile.
```

```python
# ==============================================================
# ðŸ“ˆ STRESS TESTING â€” HANDLING HIGH KURTOSIS (BFSI CONTEXT)
# ==============================================================

# High kurtosis â†’ "fat tails" â†’ more extreme values (rare but severe losses)
# Stress Testing helps financial institutions simulate and prepare
# for such extreme (tail-risk) events.

# --------------------------------------------------------------
# 1ï¸âƒ£ IMPORT LIBRARIES
# --------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, chi2_contingency

np.random.seed(42)

# --------------------------------------------------------------
# 2ï¸âƒ£ SIMULATE DATA WITH HIGH KURTOSIS
# --------------------------------------------------------------
# Using Studentâ€™s t-distribution (fat-tailed) to represent risky returns
returns = np.random.standard_t(df=3, size=1000) * 0.02  # ~2% volatility

# Compute kurtosis
kurt = kurtosis(returns)
print(f"ðŸ“Š Sample Kurtosis: {kurt:.2f}  (High â†’ Fat tails, more extreme outcomes)")

# --------------------------------------------------------------
# 3ï¸âƒ£ MONTE CARLO SIMULATION (Portfolio Stress Test)
# --------------------------------------------------------------
simulations = 10000
days = 30
initial_value = 10_000_000  # â‚¹10 million portfolio

# Randomly sample 30-day returns for each simulation
simulated_returns = np.random.choice(returns, size=(simulations, days), replace=True)
portfolio_end_values = initial_value * (1 + simulated_returns).prod(axis=1)

# Plot simulated outcomes
plt.figure(figsize=(8,5))
plt.hist(portfolio_end_values, bins=40, color='skyblue', edgecolor='black')
plt.axvline(portfolio_end_values.mean(), color='green', linestyle='--', label='Mean')
plt.axvline(np.percentile(portfolio_end_values, 5), color='red', linestyle='--', label='5th Percentile (VaR)')
plt.title("ðŸ“ˆ Monte Carlo Simulation â€” Portfolio Value After 30 Days", fontsize=13)
plt.xlabel("Portfolio Value (â‚¹)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.5)
plt.show()

# --------------------------------------------------------------
# 4ï¸âƒ£ VALUE AT RISK (VaR)
# --------------------------------------------------------------
confidence_level = 0.95
VaR_95 = initial_value - np.percentile(portfolio_end_values, (1 - confidence_level) * 100)

print(f"ðŸ’° 95% Value at Risk (VaR): â‚¹{VaR_95:,.0f}")
print("Interpretation: There's a 5% chance of losing more than this amount in 30 days.\n")

# --------------------------------------------------------------
# 5ï¸âƒ£ SCENARIO ANALYSIS
# --------------------------------------------------------------
# Test impact of specific adverse events
shock_10 = initial_value * (1 - 0.10)  # 10% sudden market drop
extreme_vol = returns * 2
extreme_sim = initial_value * (1 + extreme_vol).prod()

print("ðŸ“‰ Scenario Analysis:")
print(f" - Market Shock (10% drop): Portfolio â†’ â‚¹{shock_10:,.0f}")
print(f" - Extreme Volatility Scenario: Portfolio â†’ â‚¹{extreme_sim:,.0f}\n")

# --------------------------------------------------------------
# 6ï¸âƒ£ BFSI INTERPRETATION
# --------------------------------------------------------------
print("""
ðŸ’¡ BFSI Insights:
-----------------
âœ… High kurtosis = Fat tails â†’ Rare but severe losses possible.
âœ… Stress testing helps quantify and prepare for tail risks.

Techniques:
-----------
1ï¸âƒ£ Monte Carlo Simulation  â†’ Model thousands of possible outcomes.
2ï¸âƒ£ Value at Risk (VaR)     â†’ Quantify worst-case loss at 95% or 99% confidence.
3ï¸âƒ£ Scenario Analysis       â†’ Evaluate effect of specific market shocks.

Example:
--------
A bank observes high kurtosis in its loan loss data.
It runs Monte Carlo simulations to model extreme outcomes,
calculates 95% VaR = â‚¹1.2M, and runs a stress test for a 10% market drop.

Regulatory Relevance:
---------------------
- Used in Basel III stress testing & ICAAP reports.
- Insurance firms simulate tail events (pandemics, disasters).
- Helps ensure sufficient capital to absorb extreme losses.
""")
```

```python
# ðŸ“˜ Stress Testing for an AMC Portfolio
# =====================================
# This notebook simulates portfolio performance under stress conditions
# using Monte Carlo simulation and Value at Risk (VaR) modeling.

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1ï¸âƒ£ Portfolio Setup
# -------------------------------

# Portfolio components and their weights
assets = ['Equity', 'Bonds', 'Commodities']
weights = np.array([0.6, 0.3, 0.1])  # 60%, 30%, 10%

# Expected annual returns (in decimal form)
returns = np.array([0.10, 0.05, 0.07])

# Annualized standard deviations (volatilities)
volatility = np.array([0.18, 0.07, 0.12])

# Correlation matrix (relationship between assets)
corr_matrix = np.array([
    [1.0, 0.2, 0.3],
    [0.2, 1.0, 0.1],
    [0.3, 0.1, 1.0]
])

# Covariance matrix
cov_matrix = np.outer(volatility, volatility) * corr_matrix

# -------------------------------
# 2ï¸âƒ£ Monte Carlo Simulation
# -------------------------------

# Simulate daily returns for 1 year (252 trading days)
num_days = 252
num_simulations = 5000  # Monte Carlo runs

# Generate random correlated returns
random_returns = np.random.multivariate_normal(returns/252, cov_matrix/252, (num_simulations, num_days))

# Calculate daily portfolio returns
portfolio_returns = np.dot(random_returns, weights)

# Simulate cumulative portfolio growth (starting at â‚¹100)
portfolio_values = 100 * (1 + portfolio_returns).cumprod(axis=1)

# Plot sample portfolio paths
plt.figure(figsize=(10, 5))
plt.plot(portfolio_values[:50].T, alpha=0.3)
plt.title("Monte Carlo Simulations of Portfolio Value (â‚¹100 Base)")
plt.xlabel("Days")
plt.ylabel("Portfolio Value (â‚¹)")
plt.grid(True)
plt.show()

# -------------------------------
# 3ï¸âƒ£ Value at Risk (VaR)
# -------------------------------

# Calculate daily portfolio return distribution
daily_returns = portfolio_returns.flatten()

# 5% one-day Value at Risk (95% confidence)
VaR_95 = np.percentile(daily_returns, 5)
VaR_99 = np.percentile(daily_returns, 1)

print("ðŸ“Š Value at Risk (VaR):")
print(f" - 95% one-day VaR: {VaR_95*100:.2f}%")
print(f" - 99% one-day VaR: {VaR_99*100:.2f}%")

# Plot histogram of daily returns
plt.figure(figsize=(8,4))
plt.hist(daily_returns, bins=50, color='skyblue', edgecolor='black')
plt.axvline(VaR_95, color='red', linestyle='--', label='95% VaR')
plt.axvline(VaR_99, color='orange', linestyle='--', label='99% VaR')
plt.title("Distribution of Simulated Daily Returns")
plt.legend()
plt.show()

# -------------------------------
# 4ï¸âƒ£ Stress Scenarios
# -------------------------------

# Define custom stress scenarios (market shocks)
scenarios = {
    "Equity Crash (-30%)": np.array([-0.30, 0.00, -0.05]),
    "Interest Rate Spike (+2%)": np.array([-0.05, -0.08, 0.00]),
    "Commodity Drop (-20%)": np.array([0.00, 0.00, -0.20])
}

# Evaluate portfolio losses under each stress
print("\nâš ï¸ Stress Test Scenarios:")
for name, shock in scenarios.items():
    stress_loss = np.dot(weights, shock)
    print(f" - {name}: Portfolio impact = {stress_loss*100:.2f}%")

# -------------------------------
# 5ï¸âƒ£ Hedging Example
# -------------------------------

# Suppose AMC hedges 30% of equity exposure using index futures
hedged_weights = np.array([0.6*0.7, 0.3, 0.1])  # reduce equity risk

# Evaluate stress again
print("\nðŸ›¡ï¸ After Hedging (30% Equity Hedge):")
for name, shock in scenarios.items():
    stress_loss_hedged = np.dot(hedged_weights, shock)
    print(f" - {name}: Portfolio impact = {stress_loss_hedged*100:.2f}%")

# Comparison plot
losses = [np.dot(weights, s)*100 for s in scenarios.values()]
hedged_losses = [np.dot(hedged_weights, s)*100 for s in scenarios.values()]

df = pd.DataFrame({
    'Scenario': scenarios.keys(),
    'Unhedged Loss (%)': losses,
    'Hedged Loss (%)': hedged_losses
})

df.plot(x='Scenario', kind='bar', figsize=(8,5))
plt.title("Portfolio Loss Comparison: Before vs After Hedging")
plt.ylabel("Portfolio Loss (%)")
plt.grid(True)
plt.show()

# -------------------------------
# âœ… Summary
# -------------------------------
print("\nSummary:")
print("Monte Carlo simulation helped estimate daily portfolio risk.")
print("Value at Risk (VaR) quantified probable daily losses.")
print("Stress testing identified worst-case scenarios.")
print("Hedging reduced downside exposure effectively.")
```

# D. Percentiles, Quartiles, Deciles

```python
# ðŸ“Š Percentiles, Quartiles, and Deciles in BFSI Analytics

# These are positional measures that divide data into equal parts. They're crucial for customer segmentation, risk banding, and performance benchmarking.


## ðŸ“ Definitions

# | Concept     | Description                                                                 |
# |-------------|-----------------------------------------------------------------------------|
# | Percentile  | Divides data into 100 equal parts. The 90th percentile means 90% of values fall below it. |
# | Quartile    | Divides data into 4 equal parts: Q1 (25%), Q2 (50% = median), Q3 (75%)      |
# | Decile      | Divides data into 10 equal parts. D1 = 10th percentile, D9 = 90th percentile|



## ðŸ’¼ BFSI Use Cases

# | Metric      | BFSI Application Example                              |
# |-------------|--------------------------------------------------------|
# | Percentiles | Flagging top 5% risky transactions or high-value claims |
# | Quartiles   | Segmenting customers into low/mid/high income groups   |
# | Deciles     | Creating credit score bands for scoring models         |



## ðŸ§ª Python Implementation


import numpy as np

data = [12, 15, 21, 15, 18, 22, 15, 19, 20]

# Percentiles
p25 = np.percentile(data, 25)
p50 = np.percentile(data, 50)  # Median
p75 = np.percentile(data, 75)
p90 = np.percentile(data, 90)

# Quartiles
Q1 = p25
Q2 = p50
Q3 = p75

# Deciles
deciles = [np.percentile(data, i*10) for i in range(1, 10)]

print("25th Percentile (Q1):", Q1)
print("50th Percentile (Median):", Q2)
print("75th Percentile (Q3):", Q3)
print("90th Percentile:", p90)
print("Deciles (D1 to D9):", deciles)
```

# E. FREQUENCY DISTRIBUTIONS

```python
# ==============================================================
# ðŸ“Š FREQUENCY DISTRIBUTIONS â€” DESCRIPTIVE STATISTICS (BFSI)
# ==============================================================

# --------------------------------------------------------------
# 1ï¸âƒ£ IMPORT LIBRARIES
# --------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# For consistent random results
np.random.seed(42)

# --------------------------------------------------------------
# 2ï¸âƒ£ CREATE SAMPLE DATA
# --------------------------------------------------------------
# Example: Claim Amounts (â‚¹) or Customer Spend in BFSI
# We'll simulate continuous right-skewed data using Gamma distribution
data = np.random.gamma(shape=2, scale=2000, size=500)
df = pd.DataFrame({'Claim_Amount': data})

# --------------------------------------------------------------
# 3ï¸âƒ£ FREQUENCY DISTRIBUTION TABLE
# --------------------------------------------------------------
# Create class intervals (bins) and count observations in each
bins = [0, 1000, 2000, 3000, 4000, 5000, 7000, 10000]
labels = ['0â€“1k', '1kâ€“2k', '2kâ€“3k', '3kâ€“4k', '4kâ€“5k', '5kâ€“7k', '7kâ€“10k']

df['Range'] = pd.cut(df['Claim_Amount'], bins=bins, labels=labels, include_lowest=True)

# Frequency table
freq_table = df['Range'].value_counts().sort_index().reset_index()
freq_table.columns = ['Range', 'Frequency']

# Add relative (%) and cumulative frequencies
freq_table['Relative_Freq (%)'] = (freq_table['Frequency'] / len(df) * 100).round(2)
freq_table['Cumulative_Freq'] = freq_table['Frequency'].cumsum()
freq_table['Cumulative_%'] = freq_table['Relative_Freq (%)'].cumsum().round(2)

print("ðŸ“Š Frequency Distribution Table:\n")
display(freq_table)

# --------------------------------------------------------------
# 4ï¸âƒ£ HISTOGRAM â€” Frequency Distribution Visualization
# --------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.hist(df['Claim_Amount'], bins=bins, color='skyblue', edgecolor='black')
plt.title("ðŸ“Š Frequency Distribution of Claim Amounts", fontsize=13)
plt.xlabel("Claim Amount (â‚¹)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --------------------------------------------------------------
# 5ï¸âƒ£ RELATIVE FREQUENCY POLYGON
# --------------------------------------------------------------
# Connect midpoints of each interval to show shape of distribution
bin_midpoints = [500, 1500, 2500, 3500, 4500, 6000, 8500]
rel_freq = freq_table['Relative_Freq (%)']

plt.figure(figsize=(8,5))
plt.plot(bin_midpoints, rel_freq, marker='o', color='orange', linewidth=2)
plt.title("ðŸ“ˆ Relative Frequency Polygon", fontsize=13)
plt.xlabel("Claim Amount Midpoints (â‚¹)")
plt.ylabel("Relative Frequency (%)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --------------------------------------------------------------
# 6ï¸âƒ£ CUMULATIVE FREQUENCY (OGIVE)
# --------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(bin_midpoints, freq_table['Cumulative_%'], marker='o', color='green', linewidth=2)
plt.title("ðŸ“ˆ Cumulative Frequency (Ogive Curve)", fontsize=13)
plt.xlabel("Claim Amount Midpoints (â‚¹)")
plt.ylabel("Cumulative Percentage (%)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --------------------------------------------------------------
# 7ï¸âƒ£ INTERPRETATION (BFSI CONTEXT)
# --------------------------------------------------------------
print("""
ðŸ’¡ BFSI Insights:
-----------------
âœ… Claim Analysis:
   â€¢ Shows how many claims fall into low vs high amount ranges.
   â€¢ Helps set claim limits and detect outlier or fraudulent claims.

âœ… Customer Segmentation:
   â€¢ Identify spending ranges (low, medium, high-value customers).

âœ… Risk Modelling:
   â€¢ Examine distribution of loss amounts or credit utilization.

Example Insight:
----------------
If 70% of claims fall below â‚¹5000 â†’ most customers file small claims.
Top 10% of claims (> â‚¹7000) â†’ large, high-risk payouts.

Summary:
---------
| Type | Description | Visualization |
|------|--------------|----------------|
| Frequency Distribution | Count of observations per range | Histogram |
| Relative Frequency | % share per range | Polygon |
| Cumulative Frequency | Running total % | Ogive Curve |
""")
```

# F. Cross-tabulations and contingency tables

```python
 # ==============================================================
# ðŸ“Š CROSS-TABULATIONS & CONTINGENCY TABLES â€” DESCRIPTIVE STATISTICS (BFSI)
# ==============================================================

# --------------------------------------------------------------
# 1ï¸âƒ£ IMPORT LIBRARIES
# --------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For consistent random results
np.random.seed(42)

# --------------------------------------------------------------
# 2ï¸âƒ£ CREATE SAMPLE DATA
# --------------------------------------------------------------
# Example: Loan Portfolio Data (Categorical Variables)
# Variables:
# - Loan_Type: Type of loan (Home, Auto, Personal)
# - Default_Status: Whether customer defaulted (Yes/No)

n = 500
loan_types = np.random.choice(['Home Loan', 'Auto Loan', 'Personal Loan'], size=n, p=[0.4, 0.3, 0.3])
defaults = np.random.choice(['Yes', 'No'], size=n, p=[0.2, 0.8])

df = pd.DataFrame({
    'Loan_Type': loan_types,
    'Default_Status': defaults
})

display(df.head())

# --------------------------------------------------------------
# 3ï¸âƒ£ CREATE CROSS-TABULATION TABLE
# --------------------------------------------------------------
# Cross-tabulation shows the relationship between two categorical variables
# It counts how many observations fall into each combination of categories

crosstab = pd.crosstab(df['Loan_Type'], df['Default_Status'])
crosstab['Total'] = crosstab.sum(axis=1)
crosstab.loc['Total'] = crosstab.sum(axis=0)

print("ðŸ“Š Cross-tabulation (Contingency Table):\n")
display(crosstab)

# --------------------------------------------------------------
# 4ï¸âƒ£ ADD RELATIVE (PERCENTAGE) DISTRIBUTION
# --------------------------------------------------------------
# To interpret proportions â€” convert counts into percentages

crosstab_pct = pd.crosstab(df['Loan_Type'], df['Default_Status'], normalize='index') * 100
print("ðŸ“ˆ Row-wise Percentage Distribution:\n")
display(crosstab_pct.round(2))

# --------------------------------------------------------------
# 5ï¸âƒ£ VISUALIZATION â€” HEATMAP
# --------------------------------------------------------------
# Heatmaps provide an intuitive way to view cross-tab relationships.

plt.figure(figsize=(6,4))
sns.heatmap(crosstab_pct, annot=True, cmap="YlGnBu", fmt=".1f")
plt.title("ðŸ“Š Loan Type vs Default Status â€” Percentage Distribution", fontsize=13)
plt.ylabel("Loan Type")
plt.xlabel("Default Status")
plt.show()

# --------------------------------------------------------------
# 6ï¸âƒ£ CHI-SQUARE TEST OF INDEPENDENCE (Optional)
# --------------------------------------------------------------
# Tests whether there is a significant relationship between two categorical variables.

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(pd.crosstab(df['Loan_Type'], df['Default_Status']))

print(f"ðŸ” Chi-square Statistic = {chi2:.3f}")
print(f"p-value = {p:.4f}")
if p < 0.05:
    print("âœ… Result: Significant relationship between Loan Type and Default Status.")
else:
    print("â„¹ï¸ Result: No significant relationship between Loan Type and Default Status.")

# --------------------------------------------------------------
# 7ï¸âƒ£ INTERPRETATION (BFSI CONTEXT)
# --------------------------------------------------------------
print("""
ðŸ’¡ BFSI Insights:
-----------------
âœ… Credit Risk Analysis:
   â€¢ Cross-tabs show which loan types have higher default rates.
   â€¢ Helps in risk scoring and loan portfolio monitoring.

âœ… Marketing & Customer Segmentation:
   â€¢ Understand repayment patterns by customer segment or product type.

âœ… Fraud / Operations:
   â€¢ Identify anomalies (e.g., unusually high defaults in a specific loan category).

Example:
--------
If the 'Personal Loan' row shows 30% defaults while 'Home Loan' shows 10%, 
it indicates higher credit risk in the personal loan segment.

âœ… Summary:
------------
| Metric | Meaning | Use |
|---------|----------|-----|
| Cross-tabulation | Count of category combinations | Loan Type Ã— Default |
| Percentage Table | Proportion within each category | Compare relative risk |
| Chi-square Test | Checks independence between variables | Statistical validation |
""")
```

```python
2. Measures of Dispersion (Variability)
These show how spread out the data is:

Range: Max â€“ Min.

Variance: Average squared deviation from the mean.

Standard Deviation (SD): Square root of variance; shows average distance from mean.

Interquartile Range (IQR): Spread of the middle 50% of data (Q3 â€“ Q1).

ðŸ‘‰ Dispersion tells us whether data points are tightly clustered or widely scattered.
```

```python
3. Measures of Distribution (Shape)
These describe the overall pattern of the dataset:

Skewness: Measures asymmetry.

Positive skew â†’ long right tail.

Negative skew â†’ long left tail.

Kurtosis: Measures â€œpeakednessâ€ or flatness of distribution.

ðŸ‘‰ These help in understanding whether data follows a normal (bellâ€‘shaped) distribution or not.
```

```python
4. Frequency Distribution
Organizes data into categories or intervals and shows how often each occurs.

Represented in tables, histograms, bar charts, or pie charts.

Example: Exam scores grouped into ranges (0â€“10, 11â€“20, etc.) with counts.
```

```python
5. Graphical Representation
Visual tools make descriptive statistics intuitive:

Histograms â†’ show distribution of continuous data.

Boxplots â†’ highlight median, quartiles, and outliers.

Pie charts & bar graphs â†’ summarize categorical data.
```

```python
Why Descriptive Statistics Matter
Simplifies complexity: Large datasets become understandable.

Identifies patterns: Trends, outliers, and anomalies are visible.

Foundation for inference: Before hypothesis testing, descriptive stats provide the groundwork.

Decision-making: Businesses, researchers, and policymakers rely on these summaries for quick insights.
```
